# Results

Import packages.

```python
import pandas as pd

from pathlib import Path
```

Get a list of all the results file paths.

```python
# Create relative path to the results folder
base_dirpath = Path("../results")
input_dirpath = base_dirpath / "results"

# TODO: Fill in the file name without the extension
file_name = None

# Find all CSV files in the results directory that contain `file_name`
results_paths = list(input_dirpath.glob(f"{file_name}*.csv"))

# There should be 97 results files
print(f"Found {len(results_paths)} results files")
```

Read all of the CSV files into a single DataFrame.

```python
pd.set_option("display.max_rows", None)

df_list = []

for path in results_paths:
    df = pd.read_csv(path).tail(1)  # Read the last row of each CSV file
    df_list.append(df)

# Combine all of the dataframes into a single dataframe
df = pd.concat(df_list)

# Only keep the following columns
keep_columns = [
    "dataset",
    "model",
    "eval_metrics/MSE[mean]",
    "eval_metrics/MSE[0.5]",
    "eval_metrics/MAE[0.5]",
    "eval_metrics/MASE[0.5]",
    "eval_metrics/MAPE[0.5]",
    "eval_metrics/sMAPE[0.5]",
    "eval_metrics/MSIS",
    "eval_metrics/RMSE[mean]",
    "eval_metrics/NRMSE[mean]",
    "eval_metrics/ND[0.5]",
    "eval_metrics/mean_weighted_sum_quantile_loss",
    "domain",
    "num_variates",
]

df = df[keep_columns]

print(f"Number of rows: {len(df)}")
display(df)
```

Save the CSV file to the `all_results` directory.

```python
output_path = base_dirpath / "all_results" / file_name.replace(",", "_")
output_path.parent.mkdir(parents=True, exist_ok=True)

df.to_csv(output_path.with_suffix(".csv"), index=False)
print(f"Saved results to {output_path.with_suffix('.csv')}")
```
